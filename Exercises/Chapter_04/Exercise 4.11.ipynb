# Exercise 4.11
## With a plotting program, draw contours of the bivariate Gauss function (see next exercise for the definition of the function) for the following parameters:
## (a) $\mu_{x} = 0 $, $\mu_{y} = 0$, $\sigma_{x} = 1$, $\sigma_{y} = 1$, $\rho_{xy} = 0$
## (b) $\mu_{x} = 1 $, $\mu_{y} = 2$, $\sigma_{x} = 1$, $\sigma_{y} = 1$, $\rho_{xy} = 0.7$
## (c) $\mu_{x} = 1 $, $\mu_{y} = -2$, $\sigma_{x} = 1$, $\sigma_{y} = 2$, $\rho_{xy} = -0.7$

## Bivariate Gauss probability distribution:
\begin{equation}
P(x,y) = \frac{1}{2\pi\sigma_{x}\sigma_{y}\sqrt{1-\rho^{2}}}exp\left(-\frac{1}{2(1-\rho^{2})}\left(\frac{x^{2}}{\sigma_{x}^{2}}+\frac{y^{2}}{\sigma_{y}^{2}}-\frac{2\rho xy}{\sigma_{x}\sigma_{y}}\right)\right)
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
import math

from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

from scipy.stats import multivariate_normal
```

```python
X = np.arange(-5,5,0.05)
Y = np.arange(-5,5,0.05)

mu = np.array([0,0])
cov = np.array([[1,0],[0,1]])
mvn_a = multivariate_normal(mu,cov) #create a multivariate Gaussian object with specified mean and covariance matrix

p_a = np.zeros((len(X),len(Y)))
i = 0
j = 0
for x in X:
    j = 0
    for y in Y:
        p_a[i][j] = mvn_a.pdf(np.array([x,y]))
        j += 1
    i += 1
   

mu = np.array([1,2])
cov = np.array([[1,0.7],[0.7,1]])
mvn_b = multivariate_normal(mu,cov) #create a multivariate Gaussian object with specified mean and covariance matrix

p_b = np.zeros((len(X),len(Y)))
i = 0
j = 0
for x in X:
    j = 0
    for y in Y:
        p_b[i][j] = mvn_b.pdf(np.array([x,y]))
        j += 1
    i += 1

    
mu = np.array([1,-2])
cov = np.array([[1,-0.7],[-0.7,2]])
mvn_c = multivariate_normal(mu,cov) #create a multivariate Gaussian object with specified mean and covariance matrix

p_c = np.zeros((len(X),len(Y)))
i = 0
j = 0
for x in X:
    j = 0
    for y in Y:
        p_c[i][j] = mvn_c.pdf(np.array([x,y]))
        j += 1
    i += 1
```

```python
fig = plt.figure(figsize = (13,10))
ax = fig.gca(projection='3d')
XX, YY = np.meshgrid(X, Y)

ax.plot_surface(XX, YY, p_a, rstride=1, cstride=1, alpha=0.3) #, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(XX, YY, p_a, zdir='z', offset=0, cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_a, zdir='x', offset=0., cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_a, zdir='y', offset=0.15, cmap=cm.coolwarm)
plt.show()
```

```python
fig = plt.figure(figsize = (13,10))
ax = fig.gca(projection='3d')
XX, YY = np.meshgrid(X, Y)

ax.plot_surface(XX, YY, p_b, rstride=1, cstride=1, alpha=0.3) #, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(XX, YY, p_b, zdir='z', offset=0, cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_b, zdir='x', offset=0., cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_b, zdir='y', offset=0.15, cmap=cm.coolwarm)
plt.show()
```

```python
fig = plt.figure(figsize = (13,10))
ax = fig.gca(projection='3d')
XX, YY = np.meshgrid(X, Y)

ax.plot_surface(XX, YY, p_c, rstride=1, cstride=1, alpha=0.3) #, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(XX, YY, p_c, zdir='z', offset=0, cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_c, zdir='x', offset=0., cmap=cm.coolwarm)
#cset = ax.contour(XX, YY, p_c, zdir='y', offset=0.15, cmap=cm.coolwarm)
plt.show()
```
