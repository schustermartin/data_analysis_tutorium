```python
from __future__ import division 
import numpy as np
import matplotlib.pyplot as plt
```

## B) What is the maximum likelihood estimator for $p$?

```python
def pdf(x,p): 
    return pow(p,x)*pow((1-p), 1-x)

def d1_likelihood(x,p): 
    return -x/pow(x,2.) - (1.-x)/((1.-p)**2)
```

```python
x_stepsize = 0.01
xs = np.arange(0.+x_stepsize, 1., x_stepsize)
p_stepsize = 0.01
ps = np.arange(0.00+p_stepsize, 1., p_stepsize)
```

```python
ps = np.array([0.1,0.33,0.5,1-0.33,0.9])
plt.figure(figsize=(15, 5))
for p in ps:
    plt.plot(xs, -pdf(xs, p), label="p=%.1f" % p)
plt.title("P(x|p)")
plt.xlabel(r"$x$")
plt.grid()
plt.legend()
plt.show()
```

```python
x_stepsize = 0.1
xs = np.arange(0.+x_stepsize, 1., x_stepsize)
p_stepsize = 0.01
ps = np.arange(0.00+p_stepsize, 1., p_stepsize)
plt.figure(figsize=(15, 5))
for xi in xs:
    plt.plot(ps, pdf(xi, ps), label="x=%.1f" % xi)
plt.title("L(p)")
plt.xlabel(r"$p$")
plt.grid()
plt.legend()
plt.show()
```
