## B) What is the maximum likelihood estimator for $p$?

```python
def p(x,p): 
    return pow(p,x)*pow((1-p), 1-x)

def d1_likelihood(x,p): 
    return pow(p,x)*pow((1-p), 1-x)
```

```python
p_stepsize = 0.1
ps = np.arange(0., 1.+p_stepsize, p_stepsize)
x_stepsize = 0.01
x_max = 1.
xs = np.arange(0., x_max, x_stepsize)
```

```python
plt.figure(figsize=(15, 5))
for pi in ps:
    plt.plot(xs, p(xs, pi), label="p=%.1f" % pi)
plt.xlim(xs[1],xs[-1])
plt.grid()
plt.legend()
plt.show()
```
