Show that the likelihood based on the Binomial distribution is not properly normalized, and is therefore not a probability densitiy for $p$.

$\mathcal{L}(p) = P(r=r_0|N=N_0, p)$

$\mathcal{L}(p) = \dfrac{N!}{r!(N-r)!}p^{r}(1-p)^{N-r}$

```python
import numpy as np
import matplotlib.pyplot as plt
m = np.math
```

```python
def binomial_dest(r, N, p):
    return m.factorial(N)/(m.factorial(r)*m.factorial(N-r))*p**r*(1-p)**(N-r)
```

```python
def likelihood(p, N, r):
    return binomial_dest(r, N, p)
```

```python
# Fix N and r:
N = 1000
r = 500
```

```python
p_array = np.arange(0.,1., 0.001)
f_array = likelihood(p_array, N, r)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(p_array, f_array)
plt.show()
```

```python
# Integrate from p=0 to p=1:
a = 0.
steps = 1000
stepsize = 1./steps
for p in range(0, steps):
    p = 1.*p
    a += likelihood(p/steps, N, r)*stepsize
print(a)
```
