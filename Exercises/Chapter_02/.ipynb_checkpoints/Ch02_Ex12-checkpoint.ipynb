# Regularization

We now analyze the results in the table above using
regularization
. I.e., we
put in prior knowledge in the form of the expectation that the efficiencies
are expected to vary smoothly with energy.   One way of doing this is to
use the posterior probability density at a neighboring energy as the prior to
analyze the data.  Using the results of section 2.5.4 and starting with a flat
prior for the lowest energy point, find the posterior probability density as a
function of energy using this scheme for regularization.

```python
#Libraries that will be used for this excercise

from scipy.misc import factorial

import numpy as np #Fast and reliable array calculations and math
import matplotlib.pyplot as plt #Creating and manipulating plots

from scipy.stats import binom
from scipy.special import binom as binom_coef
# from progressbar import ProgressBar, ETA, Bar, Percentage
```

$P(p|N,r) = \dfrac{P(r|N,p)P_0(p)}{\int P(r|N,p)P_0(p)\textrm{d}p}$

```python
def posterior(p, r, N):# For the Binomial distribution and a flat Prior P_0=1, as given in the lecture.
    factor1 = factorial(N + 1.)/(factorial(r)*factorial(N-r))
    factor2 = pow(p,r) * pow((1. - p),(N - r))
    return factor1 * factor2
```

```python
# Defining numpy arrays of the given data
Energies = np.array([0.5, 1., 1.5, 2., 2.5, 3., 3.5, 4.])
Trials = np.array([100, 100, 100, 100, 100, 1000, 1000, 1000])
Successes = np.array([0,4,20,58,92,987,995,998])
# The mode of the posterior with a flat prior was in the lecture given as p* = r/N
Efficiencies_estimate = 1.0*Successes / Trials
Efficiencies_estimate
```

```python
stepsize = 1/1000.
p_values = np.arange(0,1,stepsize)#corresponds to our x axis
posterior_values = list() #will be our y axis
for energy in Energies:
     posterior_values.append(np.ones(len(p_values))) # flat priors
```

```python
def plot_posteriors_distributions(posterior_values):
    plt.figure(figsize=(15, 5))
    for idx in range(0,len(Energies)):
        plt.plot(p_values, posterior_values[idx], label="E=%.1f" % Energies[idx])
    plt.grid()
    plt.legend()
    plt.show()

plot_posteriors_distributions(posterior_values)
```

```python
for i in range(0,len(p_values)):
    posterior_values[0][i] = posterior(p_values[i] ,Successes[0], Trials[0] )

plot_posteriors_distributions(posterior_values)
```

Binomial dist.: $P(p|N,r) = \dfrac{N!}{r!(N-r)!}p^{r}(1-p)^{N-r}$

Bayes Theorem: $P(p|N,r) = \dfrac{P(r|N,p)P_0(p)}{\int P(r|N,p)P_0(p)\textrm{d}p}$

Non Flat Prior: $P(p|N,r) = \dfrac{p^{r}(1-p)^{N-r}P_0(p)}{\int\limits_0^1 p^{r}(1-p)^{N-r}P_0(p)\textrm{d}p}$

```python
def non_flat_posterior(p, N, r, prior, idx, p_array):
    posterior = 0.
    nominator = p**r*(1-p)**(N-r)*prior[idx]
    denominator = 0.
    for i in range(0,len(p_array)):
        denominator += p_array[i]**r*(1-p_array[i])**(N-r)*prior[i]
    posterior = nominator/(denominator*stepsize)    
    return posterior
```

```python
# continue with the next energy: idx=1
idx = 1
for i in range(0,len(p_values)):
    posterior_values[idx][i] = non_flat_posterior(p_values[i], Trials[idx], Successes[idx], posterior_values[idx-1], i, p_values )

plot_posteriors_distributions(posterior_values)
```

```python
# do the rest: idx >= 2
for idx in range(2,len(Energies)): 
    for i in range(0,len(p_values)):
        posterior_values[idx][i] = non_flat_posterior(p_values[i], Trials[idx], Successes[idx], posterior_values[idx-1], i, p_values )

plot_posteriors_distributions(posterior_values)
```
